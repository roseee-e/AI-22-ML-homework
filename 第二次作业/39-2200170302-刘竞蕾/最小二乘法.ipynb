```python
## 读取数据
import pandas as pd
import chardet
path = '/Users/Lenovo/Desktop/regress_data1.csv'
with open(path, 'rb') as f:  
    result = chardet.detect(f.read())  
    encoding = result['encoding'] 
data = pd.read_csv(path,encoding=encoding) ## data 是dataframe 的数据类型
  
```

```python
X_data = data.iloc[:, :-1].values  
y_data = data.iloc[:, -1].values  
```

添加偏置项

```python
X = np.c_[np.ones(X_data.shape[0]), X_data] 
```

最小二乘法求解线性回归模型

```python
W = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y_data)
w1 = W[1]  
w0 = W[0]
```

```python
y_pred = X.dot(W) 
```

```python
plt.figure(figsize=(8, 6))  
plt.scatter(X_data, y_data, color='blue', label='原始数据')  
plt.plot(X_data, y_pred, color='red', label='线性回归拟合线')  
plt.xlabel('人口')  
plt.ylabel('收益')  
plt.legend()  
plt.title('线性回归模型预测结果')  
plt.show()
```
